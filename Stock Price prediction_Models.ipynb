# 3. Comparison of performance of models
    - Stock example : "NFLX"

```python
model_names = ["Linear", "SVM", "AdaBoost", "ExtraTrees", "RandomForest", "XGB", "LGMB", "MLP", "LSTM", "GRU"]
input_ex = "NFLX"

ans = dict()
acc = dict()
for model in model_names:
    print(model)
    ai = AI_Stock(input_ex, model_name=model)
    print("\n ==============<< {} Models Results >>============== \n".format(model))
    if model == "LSTM":
        %time acc[model] = ai.rnn_predict()
    elif model == "GRU":
        %time acc[model] = ai.rnn_predict()
    else:
        %time ans[model], acc[model] = ai.short_term_predict()
```
